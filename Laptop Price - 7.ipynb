# Data Visualization 

```python
# Showing the Line Plots

sns.set(style = 'whitegrid')
plt.plot(laptop['Ram'], laptop['OS'], marker = '*', ms = 15, mec = 'r', color = "black", linestyle = '--', lw = 0.5)
plt.title('Ram Vs OS', color = 'blue')
plt.xlabel("Ram", color = 'navy')
plt.ylabel("OS", color = 'y')
plt.show()
```

```python
# Showing the Bar Plots

plt.bar(laptop['Screen'], laptop['PrimaryStorage'], width = 0.5, color = 'silver', edgecolor = 'black')
plt.title('Comparision b/w Screen and PrimaryStorage', color = 'red')
plt.xlabel("Screen", color = 'r')
plt.ylabel("PrimaryStorage", color = 'b')
plt.show()
```

```python
# Showing the Histogram - 1

plt.hist(laptop['Weight'], bins = 20, histtype = 'barstacked', color = 'g', edgecolor = 'black')
plt.title("Showing the Weight histogram", color = 'y')
plt.xlabel("Weight",color = 'indigo')
plt.ylabel("Count", color = 'teal')
plt.show()
```

```python
# Showing Scatter Plot

plt.scatter(laptop['Inches'], laptop['Weight'], marker = "D", c = 'c', s = 10, alpha = 0.50 )
plt.xlabel("Inches", color = 'g')
plt.ylabel("Weight", color = 'b')
plt.title("Ploting the point's b/w Inches and Weight ", color = 'orange')
plt.colorbar()
plt.show()
```

```python
# Shwoing Pie - Chart

plt.figure(figsize = (15,10))
sakku = laptop[['Ram','ScreenW','PrimaryStorage']].value_counts().head(10)
c = ['orange','yellow','skyblue','pink','black','lightgreen','silver','red','blue','green']
plt.pie(sakku, labels = sakku.index, autopct = '%1.2f%%', startangle = 15, shadow = True , colors = c)
plt.title("Combination of Ram, ScreenW and PrimaryStorage", color = 'brown')
plt.legend(title = "10 Pair's")
plt.show()
```

```python
# Showing lineplot()

plt.figure(figsize = (8,7))
sns.lineplot(data = laptop, x = 'Weight', y = 'Price_euros', marker = 'D', mec = 'black',color = 'skyblue')
plt.xlabel("Weight", color = 'Indigo')
plt.ylabel("Price_euros", color = 'Purple')
plt.title("Difference b/w Weight and Price_euros", color = 'r')
plt.grid(False)
plt.show()
```

```python
# Showing barplot()

plt.figure(figsize = (10,7))
sns.barplot(data = laptop, x = 'ScreenW', y = 'CPU_freq', hue = 'PrimaryStorageType', width = 1, capsize = 0.1)
plt.xlabel("ScreenW", color = 'red')
plt.ylabel("CPU_freq", color = 'Teal')
plt.title("Comparision b/w ScreenW and CPU_freq", color = "Maroon")
plt.show()
```

```python
# Showing scatterplot()

plt.figure(figsize = (9,8))
sns.scatterplot(data = laptop, x = 'Inches', y = 'Price_euros', hue = 'GPU_company', marker = 'D' )
plt.xlabel("Inches", color = 'Gray')
plt.ylabel("Price_euros", color = 'Navy')
plt.title("Plotting the scrattering points b/w Inches and Price_euros ", color = 'green')
plt.show()
```

```python
# Showing boxplot()

sns.set(style = 'whitegrid')
plt.figure(figsize = (8,7))
sns.boxplot(data = laptop, x = 'Ram', y = 'Weight', color = 'Coral', width = 1, gap = 0.2)
plt.xlabel("Ram", color = 'r')
plt.ylabel("Weight", color = 'teal')
plt.title("Comapring Distribution b/w Ram and Weight", color = 'navy') 
plt.show()
```

```python
# Showing violinplot()

sns.set(style = 'whitegrid')
sns.violinplot(data = laptop, x = 'Ram', y = 'Inches', color = 'teal', cut = 5, width = 1, gridsize = 18 )
plt.xlabel("Ram", color = 'r')
plt.ylabel("Inches", color = 'b')
plt.title("Identifying the Density Information b/w Ram and Inches", color = 'Crimson')
plt.show()
```

```python
# Showing histplot()

sns.set(style = 'whitegrid')
plt.figure(figsize = (7,6))
sns.histplot(data = laptop, x = 'Price_euros', hue = 'CPU_company', kde = True, bins = 15)
plt.xlabel("Price_euros", color = 'indigo')
plt.ylabel("Count", color = 'Olive')
plt.title("Continuous Histogram of Price_euros ",color = 'green')
plt.show()
```

```python
# Showing countplot()

plt.figure(figsize = (11,7))
sns.countplot(data = laptop, x = 'OS', color = 'gold', width = 0.5)
plt.xlabel("OS", color = 'Maroon')
plt.ylabel("Count", color = 'orange')
plt.title("Comparing Frequency in the OS", color = 'Navy')
plt.show()
```

```python
# Showing pairplot()

sns.pairplot(data = laptop, hue = 'Ram')
#plt.suptitle("Showing the pairwise relationships in a data", color = 'brown') 
plt.show()
```

```python
# Showing heatmap()

plt.figure(figsize = (9,8))
sakku = laptop[['Inches', 'Ram', 'Weight', 'Price_euros', 'ScreenW', 'CPU_freq', 'PrimaryStorage']].corr()
sns.heatmap(sakku, annot = True, cmap = 'plasma')
plt.show()
```
