# Data Visualization

```python
# Showing line plot()

plt.plot(movie['Year'], movie['MetaScore'], marker = 'D', ms = 6, mec = 'c', color = 'black', linestyle = '--')
plt.xlabel("Year", color = 'red')
plt.ylabel("MetaScore", color = 'b')
plt.title(" Year Vs MetaScore", color = 'green')
plt.show()
```

```python
# Showing barplots()

plt.figure(figsize = (7,6))
plt.bar(movie['MetaScore'], movie['Certificates'], width = 0.5, color = 'green', edgecolor = 'black')
plt.xlabel("MetaScore", color = 'orange')
plt.ylabel("Certificates", color = 'purple')
plt.title("Comparision b/w MetaScore and Certificates", color = 'maroon')
plt.show()
```

```python
# Showing histgrom()

sns.set(style = 'whitegrid')
plt.hist(movie['IMDb Rating'], bins = 20, color = 'green', edgecolor = 'black')
plt.xlabel('IMDb Rating', color = 'black')
plt.ylabel('Count', color = 'navy')
plt.title("Showing the IMDb Rating Histogram of Continuous", color = 'indigo')
plt.show()
```

```python
# Showing the Scatterplot()

plt.scatter(movie['Year'], movie['MetaScore'], marker = '*', color = 'black', s = 65, edgecolor = 'b', alpha = 0.5)
plt.xlabel('Year', color = 'brown')
plt.ylabel('MetaScore', color = 'teal')
plt.title("Plotting the points b/w Year and MetaScore", color = 'crimson')
plt.colorbar()
plt.show()
```

```python
# Showing pie-chart()

plt.figure(figsize = (11,13))
sakku  = movie[['Year', 'Duration (minutes)']].value_counts().head(10)
c = ['green','purple','pink','brown','gray','lime','teal','olive','gold','silver']
plt.pie(sakku, labels = sakku.index, autopct = '%1.1f%%', startangle = 10, shadow = True , colors = c)
plt.title("Combination of Year and Duration (minutes)", color = 'brown')
plt.legend(title = "10 Pair's")
plt.show()
```

```python
# Showing lineplot()

sns.lineplot(data = movie, x = 'Year', y = 'MetaScore', marker = '*', ms = 10, mec ='r', color = 'indigo' )
plt.xlabel('Year', color = 'b')
plt.ylabel('MetaScore', color = 'brown')
plt.title("Difference b/w Year and MetaScore", color = 'red')
plt.show()
```

```python
# Showing barplot()

#plt.figure(figsize = (10,8))
sns.barplot(data = movie, x = 'MetaScore', y = 'Certificates', color = 'brown', width = 0.5, capsize = 0.6)
plt.xlabel('MetaScore', color = 'teal')
plt.ylabel('Certificates', color = 'coral')
plt.title("Comparision b/w MetaScore and Certificates", color = 'olive')
plt.show() 
```

```python
# Showing scatterplot()

plt.figure(figsize = (9,8))
sns.scatterplot(data = movie, x = 'IMDb Rating', y = 'Genre', marker = 'D', hue ='Certificates')
plt.xlabel('IMDb Rating', color = 'r')
plt.ylabel('Genre', color = 'blue')
plt.title("Plotting the scattering points b/w IMDb Rating and Genre", color = 'maroon')
plt.show()
```

```python
# Shwoing histplot()

sns.histplot(data = movie, x = 'Duration (minutes)', hue = 'Genre', kde = True, bins = 20, binwidth = 10)
plt.xlabel('Duration (minutes)', color = 'lime')
plt.ylabel('Count', color = 'olive')
plt.title("Continuous Histogram of Duration (minutes)", color = 'crimson')
plt.show()
```

```python
# Showing boxplot()

sns.boxplot(data = movie , x = 'IMDb Rating', y = 'Genre', color = 'brown', width = 0.5, gap = 0.2 )
plt.xlabel('IMDb Rating', color = 'navy')
plt.ylabel('Genre', color = 'gray')
plt.grid(True)
plt.title("Comparing Distribution b/w IMDb Rating and Genre", color = 'green')
plt.show()
```

```python
# Showing countplot()

plt.figure(figsize = (10,8))
sns.countplot(data = movie, x = 'Genre', hue = 'Certificates', edgecolor = 'c')
plt.xlabel('Genre' , color = 'green')
plt.ylabel('Count', color = 'black')
plt.title("Comparing Frequency in the Genre", color = 'navy')
plt.show()
```

```python
# Showing violinplot()

plt.figure(figsize = (10,7))
sns.violinplot(data = movie, x = 'Genre', y = 'IMDb Rating', color = 'olive', cut = 5, width = 0.9, gridsize = 20 )
plt.xlabel("Genre", color = 'r')
plt.ylabel("IMDb Rating", color = 'b')
plt.title("Identifying the Density Information b/w Genre and IMDb Rating", color = 'Crimson')
plt.show()
```

```python
# Showing pairplot()

sns.pairplot(data = movie, hue = 'Year')
#plt.suptitle("Showing the pairwise relationships in the data")
plt.show()
```

```python
#Showing heatmap()

sakku = movie[['IMDb Rating','Year', 'MetaScore','Duration (minutes)']].corr()
sns.heatmap(sakku, annot = True, cmap = 'magma')
plt.show()
```
