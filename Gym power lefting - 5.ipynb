# Data Visualiazation

```python
# Showing the line plot

plt.plot(Gym['Age'], Gym['Weight Class'], marker = 'D', color = 'skyblue', ms = 8, mec = 'black', linestyle = '--', lw = 0.8)
plt.xlabel("Age", color = 'orange')
plt.ylabel("Weight Class", color = 'purple')
plt.title(" Age Vs Weight Class", color = 'red')
plt.show()
```

```python
# Showing barplot()

#plt.figure(figsize = (10,8))
plt.bar(Gym['Age'], Gym['Lift Type'], width = 0.50, color = 'c', edgecolor = 'black' )
plt.xlabel("Age", color = 'brown')
plt.ylabel("Lift Type", color = 'teal')
plt.title('Comparision b/w Age and Lift Type', color = 'navy')
plt.show()
```

```python
# Showing Histogram

plt.hist(Gym['Age'], bins = 35, width = 2, histtype = 'barstacked', color = 'black', edgecolor = 'c')
plt.xlabel("Age", color = 'olive')
plt.ylabel("Counts", color = 'brown')
plt.title("Continuous of the Age", color = 'maroon')
plt.show()
```

```python
# Showing the scatterplot()

plt.scatter(Gym['Age'], Gym['Weight Class'], marker = 'D', c = 'indigo', s = 15, alpha = 0.50)
plt.xlabel("Age", color = 'gray')
plt.ylabel("Weight Class", color = 'navy')
plt.title("Ploting the points b/w Age and Weight Class", color = 'blue')
plt.colorbar()
plt.show()
```

```python
# Showing Pie - Chart

plt.figure(figsize = (10,10))
sakku = Gym[['Age', 'Weight Class']].value_counts().head(8)
c = ['coral', 'crimson', 'indigo', 'lime', 'olive', 'teal', 'gray', 'skyblue', 'orange', 'navy']
plt.pie(sakku, labels = sakku.index, autopct ='%1.2f%%', shadow = True, colors = c)
plt.legend(title = "10 Pair's")
plt.show()
```

```python
# Showing lineplot()

plt.figure(figsize = (8,7))
sns.lineplot(data = Gym, x = 'Age', y = 'Weight Class', hue = 'Lift Type', marker = 'D', mec = 'b', size = 10)
plt.xlabel("Age", color = 'r')
plt.ylabel("Weight Class", color = 'navy')
plt.title("Difference b/w Age and Weight Class", color = 'brown')
plt.show()
```

```python
# Showing barplot()

#plt.figure(figsize = (8,7))
sns.barplot(data = Gym, x = 'Amount Lifted (kg)', y = 'Lift Type', color = 'pink', width = 0.4, capsize = 0.10, gap = 0.10)
plt.xlabel('Amount Lifted (kg)', color = 'maroon')
plt.ylabel('Lift Type', color = 'brown')
plt.title("Comparision b/w Amount Lifted (kg) and Lift Type", color = 'crimson')
plt.show()
```

```python
# Showing Scatterplot()

sns.scatterplot(data = Gym, x = "Age", y = "Amount Lifted (kg)", marker = "D", color = 'teal', alpha = 0.50 )
plt.xlabel("Age", color = 'gray')
plt.ylabel("Amount Lifted (kg)", color = 'navy')
plt.title("Plotting the scattering points b/w Age and Amount Lifted (kg)", color = 'indigo')
plt.show()
```

```python
# Showing histplot()

plt.figure(figsize = (8,7))
sns.set(style = 'whitegrid')
sns.histplot(data = Gym, x = 'Age', hue = 'Lift Type', bins = 15, kde = True)
plt.xlabel("Age", color = 'red')
plt.ylabel("Count", color = 'blue')
plt.title("Continuous Histogram of Age", color = 'green')
plt.show()
```

```python
# Showing boxplot()

sns.boxplot(data = Gym, x = 'Lift Type', y = 'Amount Lifted (kg)', color = 'brown', width = 0.5, gap = 0.5,)
plt.xlabel('Lift Type', color = 'olive')
plt.ylabel('Amount Lifted (kg)', color = 'teal')
plt.grid(True)
plt.title("Comparinig Distribution b/w Lift Type and Amount Lifted (kg)", color = 'navy')
plt.show()
```

```python
# Showing violinplot()

sns.violinplot(data = Gym,  x = 'Weight Class', y = 'Amount Lifted (kg)', color = 'gray', width = 1, cut = 5,  gridsize = 20)
plt.xlabel("Weight Class", color = 'teal')
plt.ylabel("Amount Lifted (kg)", color = 'maroon')
plt.title("Identifying the Density Information b/w Weight Class and Amount Lifted (kg)", color = 'coral') 
plt.show()
```

```python
# Showing countplot()

sns.countplot(data = Gym, x = 'Lift Type', color = 'indigo', width = 0.5)
plt.xlabel("Lift Type", color = 'Maroon')
plt.ylabel("count", color = 'orange')
plt.title("Comparing Frequency in the Lift Type", color = 'Navy')
plt.show()
```

```python
# Showing pairplot() -  1

sns.pairplot(Gym)
plt.show()
```

```python
# Showing pairplot() -  2

sns.pairplot(Gym, hue = "Age")
plt.show()
```

```python
# Showing heatmap()

sakku = Gym[['Age', 'Amount Lifted (kg)']].corr()
sns.heatmap(sakku, annot = True, cmap = 'plasma')
plt.show()
```
