# Data Visualization

```python
# Showing the Line Plot

plt.plot(data['Age'], data['PurchaseStatus'], marker = '*', ms = 15, mec = "r", linestyle  = '--', lw = 1)
#plt.xlable = 'Age'
#plt.ylabel= 'PurchaseStatus'
plt.title('Age vs. PurchaseStatus', color = 'orange')
plt.show()
```

```python
# Showing Bar Plot

plt.bar(data['Age'], data['TimeSpentOnWebsite'], width = 0.2, color = 'r')
plt.xlable=('Age')
plt.ylabel=('TimeSpentOnWebsite')
plt.title('Age vs TimeSpentOnWebsite')
plt.show()
```

```python
# Showing Histogram 

plt.figure(figsize=(10,6))
plt.hist(data['NumberOfPurchases'], histtype = 'bar', width = 0.9, bins = 500, bottom = "CPD", color  = "c", edgecolor = 'k')
plt.xlabel('NumberOfPurchases')
#plt.ylabel('Frequency')
plt.title('NumberOfPurchases', color = 'r')
plt.show()
```

```python
# Showing the Scatter Plot

plt.scatter(data['AnnualIncome'], data['NumberOfPurchases'], marker = '*', c = 'c', s = 50, alpha = 0.5 )
plt.scatter(data['Age'], data['Gender'], marker = 'D', c = 'b', s = 10)
plt.title('    Age/Gender  Vs   AnnualIncome/NumberOfPurchases', color = 'y')
plt.colorbar()
plt.show()
```

```python
# Showing Pie - Chart

a = data[[ 'ProductCategory', "PurchaseStatus"]].value_counts()
plt.figure(figsize=(12,20))
c = ['red', 'green', 'black','orange','yellow','skyblue','pink','blue','lightgreen','silver']
plt.pie(a, labels = a.index, autopct = '%1.2f%%', startangle = 100, shadow = True, colors = c )
plt.title('Combination of ProductCategory and PurchaseStatus', color = 'g')
plt.legend(title = "10 Pair's")
plt.show()
```

```python
# Showing lineplot()

sns.lineplot( data = data, x = 'Age', y = 'NumberOfPurchases', size = 15, marker = 'D',sort = True, color = 'black', mec = 'c')
plt.xlabel("Age", color = 'b')
plt.ylabel("NumberOfPurchases", color = 'Teal')
plt.title("Difference b/w Age and NumberOfPurchases", color = 'r')
plt.show()
```

```python
# Showing barplot()

plt.figure(figsize = (9,8))
sns.barplot( data = data, x = 'NumberOfPurchases', y = 'DiscountsAvailed', hue = "PurchaseStatus", gap = 0)
plt.xlabel("NumberOfPurchases", color = 'orange')
#plt.ylabel("DiscountsAvailed", color = 'r')
plt.title("Comparision b/w NumberOfPurchases and DiscountsAvailed", color = 'b')
plt.show()
```

```python
# Showing scatterplot()

plt.figure(figsize = (10,9))
sns.scatterplot(data = data, x = 'Age', y = 'AnnualIncome', marker = 'D', hue = 'DiscountsAvailed')
plt.xlabel("Age", color = 'r')
#plt.ylabel("AnnualIncome", color = 'b')
plt.title("Plotting the scattering points b/w Age and AnnualIncome", color = 'brown')
plt.show()
```

```python
# Showing histplot()

sns.histplot(data = data, x = 'Age', kde = True, hue = "Gender", bins = 15, binwidth = 2)
plt.xlabel("Age", color = 'violet')
plt.title("Continuous Histrogram Of Age", color = 'green')
plt.show()
```

```python
# Showing boxplot()

sns.boxplot(data = data, x = 'NumberOfPurchases', y = 'DiscountsAvailed', color = 'pink', width = 0.5)
plt.xlabel("NumberOfPurchases", color = 'r')
plt.title("Comparing Distribution b/w NumberOfPurchases and DiscountsAvailed", color = 'Teal')
plt.show()
```

```python
# Showing violinplot()

sns.violinplot(data = data, x = "NumberOfPurchases", y = "DiscountsAvailed", color = 'c', width=0.9)
plt.xlabel("NumberOfPurchases", color = 'Magenta')
plt.title("Identifying the Density Information b/w NumberOfPurchases and DiscountsAvailed", color = 'Navy')
plt.show()
```

```python
# Showing countplot()

sns.countplot(data = data, x = 'Gender', color = 'c')
plt.xlabel("Gender", color = 'Indigo')
plt.title("Comparing Frequencies in the Gender", color = 'Purple')
plt.show()
```

```python
# Showing pairplot()

sns.pairplot(data = data, hue = 'Gender')
#plt.title("Visualizing Relationship of Gender", color = 'Brown')
plt.show()
```

```python
# Showing heatmap()

plt.figure(figsize = (10,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', lw = 0.5)
plt.show()
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(data.corr(), annot=True, cmap='plasma')
plt.show()
```
